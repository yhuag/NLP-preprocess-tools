```python
from lingpy import *
import pandas as pd
import numpy as np
```

```python
file_names = ["french.txt", "italian.txt", "portuguese.txt", "spanish.txt"]
```

```python
def loadFile2Df(file_name):
    data = pd.read_csv(file_name, sep="____", header=None,engine='python')
    data.columns = ["w1", "w2", "cognate"]
    return pd.DataFrame(data)
```

```python
def detectCognate(df):
    total_row = len(df.index)
    total_correct = 0
    for index, row in df.iterrows():
        cognate_pred = "cognates" if turchin(row['w1'], row['w2']) == 0 else "non_cognates"
        correctness = (cognate_pred == row["cognate"])
        if correctness:
            total_correct = total_correct + 1
#             print("Correct: ", row["w1"], row["w2"], row["cognate"], cognate_pred)
#         else:
#             print("NOT Correct: ", row["w1"], row["w2"], row["cognate"], cognate_pred)
    return total_correct/total_row
```

```python
df = loadFile2Df("french_cognate.txt")
acc = detectCognate(df)
print(acc)
```

```python
for file_name in file_names:
    df = loadFile2Df(file_name)
    acc = detectCognate(df)
    print(file_name, acc)
```

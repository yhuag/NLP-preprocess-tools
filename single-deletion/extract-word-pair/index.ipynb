```python
import pandas as pd
import numpy as np
```

```python
final_ans = []
```

```python
def is_single_deletion(w1, w2, is_first):
    if abs(len(w1) - len(w2)) != 1 and is_first:
        return False
    
    cnt = 0
    for (c1, c2) in zip(w1, w2):
        if c1 != c2:
            if not is_first:
                return False
            if len(w1) > len(w2):
                return is_single_deletion(w1[cnt+1:], w2[cnt:], False)
            elif len(w1) < len(w2):
                return is_single_deletion(w1[cnt:], w2[cnt+1:], False)
            
        # counter increment
        cnt = cnt + 1
    return True
```

```python
is_single_deletion("rosmarin", "romarin", True)
```

```python
def check(file_name):
    # Load text files
#     data = pd.read_csv(file_name, sep="____", header=None)
    data = pd.read_csv(file_name, delim_whitespace=True, header=None, error_bad_lines=False)
#     data.columns = ["w1", "w2", "cognate"]
    data.columns = ["w1", "w2"]
    
    # Load as dataframe
    df = pd.DataFrame(data)
    
    # Pre-filtering
#     df_filtered = df[df.cognate == "cognates"]
    df_filtered = df
    df_filtered = df_filtered[(df_filtered.w1.str.len() <= 8) & (df_filtered.w2.str.len() <= 8)]
    df_filtered = df_filtered[abs(df_filtered.w1.str.len() - df_filtered.w2.str.len()) == 1]
    
    # Check isSingle
    for index, row in df_filtered.iterrows():
        isSingle = is_single_deletion(row['w1'], row['w2'], True)
        print(row['w1'], row['w2'], isSingle)
        if isSingle:
            final_ans.append((row['w1'], row['w2']))
```

```python
file_names = ["french.txt", "italian.txt", "portuguese.txt", "spanish.txt", "turkish.txt"]
# file_names = ["french_test.txt", "french_train.txt", "italian_test.txt", "italian_train.txt", "portuguese_test.txt", "portuguese_train.txt", "spanish_test.txt", "spanish_train.txt"]
```

```python
for file_name in file_names:
    check(file_name)
```

```python
# Strict answers (deletion error in the middle)
strict_ans = []
for (w1, w2) in final_ans:
    if w1[-1] == w2[-1] and w1[0] == w2[0]:
        strict_ans.append((w1,w2))
```

```python
# Get perfect examples
for (w1, w2) in strict_ans:
    # Make m1 always the longest word
    if len(w2) > len(w1):
        w3 = w1
        w1 = w2
        w2 = w3
        
    # Catch the changed index
    idx = 0
    for (c1,c2) in zip(w1,w2):
        if c1 != c2:
            break
        else:
            idx = idx + 1
            
    # Update mid index
    if (len(w1)%2)==0:
        id_mid = int(len(w1)/2 - 1)
    else:
        id_mid = int((len(w1)+1)/2 - 1) 
        
    # Final check
    if w1[id_mid] == w1[id_mid+1] and w1[id_mid] == w1[idx]:
        print(w1, w2)
    
```
